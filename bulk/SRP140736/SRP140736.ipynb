#### anat entity
anat entity according to methods from paper: 20 g of longissimus dorsi (LD) muscle sample from the left side

longissimus dorsi isn't a UBERON term specifically, longissimus thoracis has the synonym longissimus dorsi and this is the term we use

annotating with longissimus thoracis muscle (UBERON:0001401)
- annotation status is perfect match
- also adding anatBiologicalStatus as partial sampling

```python
library.loc[:,'anatId'] = 'UBERON:0001401'
library.loc[:,'anatName'] = 'longissimus thoracis muscle'
library.loc[:,'anatAnnotationStatus'] = 'perfect match'
library.loc[:,'anatBiologicalStatus'] = 'partial sampling'
```

```python
display_df(library)
```

#### stage
anat stage according to paper: Eighty newborn lambs with near-average birth weights were euthanized **after birth** by carbon dioxide inhalation followed by exsanguination

[sheep developmental stages reference](https://genomics.senescence.info/species/entry.php?species=Ovis_aries)
we have no "birth" stage or anything super specific in that time period
- lastest pre-birth term we have is "day 21 and over embryonic stage", part of the organogenesis stage
- gestation is approx 146 days 
- shortest and most specific post birth stage is "suckling stage": Sheep developmental stage that covers the period from birth until about 6 months old, at which point individuals wean.

annotating with suckling stage and missing child term

```python
library.loc[:,'stageId'] = 'OariDv:0000004'
library.loc[:,'stageName'] = 'suckling stage'
library.loc[:,'stageAnnotationStatus'] = 'missing child term'
```

```python
display_df(library)
```

#### sex, strain, genotype, speciesId

sex
- update to M/F 

strain
- Hu sheep is not present in [uniprot strain list](https://ftp.uniprot.org/pub/databases/uniprot/current_release/knowledgebase/complete/docs/strains)
- Hu sheep is found on the [OK State livestock breed](https://breeds.okstate.edu/sheep/hu-sheep.html) site
- update to just "Hu" instead of "Hu sheep"

genotype
- no info given, leaving blank is correct

species ID
- 9940 is correct for Ovis aries

```python
library.loc[library["sex"] == "male", "sex"] = "M"
library.loc[library["sex"] == "female", "sex"] = "F"
library.loc[:,'strain'] = 'Hu'
```

```python
display_df(library)
```

#### protocol, globin, replicates

protocol: per methods, NEBNext Ultra RNA Library Prep Kit for Illumina with poly(A) mRNA isolation
- protocol = NEBNext Ultra RNA Library Prep Kit
- protocolType = polyA
- RNASelection = full_length

globin reduction: not applicable

replicates: these are biological replicates, not technical replicates

```python
library.loc[:,'protocol'] = 'NEBNext Ultra RNA Library Prep Kit'
library.loc[:,'protocolType'] = 'full_length'
library.loc[:,'RNASelection'] = 'polyA'
```

```python
display_df(library)
```

#### sample age, pato, physiological status

no specific details given on sample age (just newborn), no additional detail to add here

no physiologicalStatus or PATO to add here


#### condition

condition: control (C), control supplemented with 16 mg (F16) or 32 mg (F32) effective rumen-protected folic acid per kilogram dry matter intake (DMI)
- control is no folic acid
- folate 1 is 16 mg folic acid per kg DMI
- folate 2 is 32 mg folic acid per kg DMI

historically at bgee, addition of dietary supplements (especially things that can be found in environment) can still considered healthy wild type

updating condition column to be more specific about the amount of folic acid given
- control (no folic acid dietary supplement)
- folate 1 (dietary supplement 16 mg folic acid per kg dry matter intake)
- folate 2 (dietary supplement 32 mg folic acid per kg dry matter intake)

```python
library.loc[library["condition"] == "control", "condition"] = "control (no folic acid dietary supplement)"
library.loc[library["condition"] == "folate 1", "condition"] = "folate 1 (dietary supplement 16 mg folic acid per kg dry matter intake)"
library.loc[library["condition"] == "folate 2", "condition"] = "folate 2 (dietary supplement 32 mg folic acid per kg dry matter intake)"
```
